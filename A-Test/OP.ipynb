```python
import openai
from openai import OpenAI

# 设置基本参数
client = OpenAI(
    base_url="https://api.openai-proxy.org/v1",
    api_key="sk-2px7Nzng8D0OOWC1VGAo209S05XrOGmoswZZYhphdEaOvjMk"
)
model_name = 'gpt-4o-mini'

# 定义调用函数
def gpt_retry(**kwargs):
    return client.chat.completions.create(**kwargs)

def gpt_handler(prompt, model, sc=1, T=0.8):
    response = gpt_retry(
        model=model,
        messages=[
            {"role": "system", "content": "Follow the given examples and answer the question, you should return the COT following the Answer"},
            {"role": "user", "content": prompt}
        ],
        n=sc,  # number of outputs per question
        temperature=T,
        timeout=60  # 修正参数名
    )
    # 修正响应处理方式
    ans_model = [choice.message.content for choice in response.choices]
    return ans_model

if __name__ == '__main__':
    results = gpt_handler("告诉我一些关于蒙特卡洛方法相似的方法", model_name, 3)
    for i, result in enumerate(results, 1):
        print(f"回答 {i}:")
        print(result)
        print("-" * 50)
```
